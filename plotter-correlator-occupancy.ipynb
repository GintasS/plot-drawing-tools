```python
def get_quantiles_values(quantiles, histo):
    res=np.zeros(quantiles.shape)
    n = histo.GetQuantiles(quantiles.shape[0], res, quantiles)
    for id,quantile in enumerate(quantiles):
        print ('quantile {}% = {:.2f}'.format(quantile*100, res[id]))
    return res
```
